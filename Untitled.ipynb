```python
sql = '''
SELECT
    histogram(flux2mag(forced.i.psfflux_flux), range => (10, 30), bins => 50)
FROM
    pdr2_dud
'''

%time res = sqlclient.post_sql_with_tqdm(sql)

from matplotlib import pyplot

hist, bins = res.target_list[1][0]
pyplot.plot(bins[1:], hist)
pyplot.grid()
```
